```python
import math
import random

class Point:
    def __init__(self, i, j, ic, jc, k, sigma):
        self.i = i
        self.j = j
        self.ic = ic
        self.jc = jc
        self.k = k
        self.sigma = sigma

    def dist(self):
        return ((self.i-self.ic)**2 + (self.j-self.jc)**2)**0.5

    def pab(self):
        return math.exp(-self.dist()**2/(self.k*self.sigma))

    def status(self):
        return 1-int(random.random()+1-self.pab())
```

```python
p = Point(3, 4, 0, 0, 1, 1)
p.dist()
random.random()
```

```python
p.pab()

import matplotlib.pyplot as plt
import numpy as np

# Initialize the list to store the pab values
pab_values = []

# Define the range of k values
k_values = np.arange(1, 1500, 1)

# Calculate the pab value for each k
for k in k_values:
  p = Point(4, 3, 0, 0, k, 6)
  pab_values.append(p.status())

# Plot the histogram
plt.hist(pab_values, bins=20, edgecolor='black')
plt.xlabel('pab')
plt.ylabel('Frequency')
plt.title('Distribution of pab while k varies')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create a 30x50 picture
plt.figure(figsize=(5, 5))

dn=1
dm=1
ic=10
jc=10
k=200
sigma=5

# Create and draw the points
for i in np.arange(0, 100, dm):
    for j in np.arange(0, 100, dn):
        p = Point(i, j, ic, jc, k, sigma)
        plt.plot(p.pab())
        plt.plot(p.status())
        plt.plot(p.dist())
        if p.status() == 1:
            plt.scatter(j, i, s=0.1, color='r')
                      # Note that j is the x-coordinate and i is the y-coordinate

# Set the limits of the axes
#plt.xlim(0, 50)
#plt.ylim(0, 30)

# Show the picture
plt.show()
```
